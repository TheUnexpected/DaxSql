```python
import pyodbc
import pandas as pd

#Conexion a base de datos
#=================================================
def conexion(Driver):
    server = 'lnmxvmsqlqa02' 
    database = 'HDI_DWH'
    
    try:
        if Driver== "WINDOWS":
            #Driver Windows: MAQUINA VIRTUAL
            cnxn = pyodbc.connect('DRIVER={SQL Server};SERVER='+server+';DATABASE='+database+';Trusted_Connection=yes;')
        elif Driver == "LINUX":
            #Driver Linux: MobaXTerm
            cnxn = pyodbc.connect('DRIVER=FreeTDS;SERVER='+server+';PORT=1433;DATABASE='+database+';UID=test;PWD=test')
    except Exception as e:
        print('1. Error en la conexion, favor de verificar')
    else:
        print('1. Connection to server: ' + database + ', succesfull')
    return cnxn
#-------------------------------------------------

#Consulta de query a base de datos 
#=================================================
def conectaTablaBD(query, conexion):
    data= ''
    try:
        sql = query
        data = pd.read_sql(sql, conexion)
    except Exception as e:
        print('2. Error en la conexion, favor de verificar')
    else:
        print('2. Lectura de DataBase: correcta')
    return data
#-------------------------------------------------

#Leer query de un archivo .sql
#=================================================
def OpenFile(File):
    contenido=''
    try:
        with open(File, 'r', encoding='utf-8') as archivo:
            # Lee el contenido del archivo
            contenido = archivo.read()
    except: 
        with open(File, 'r') as archivo:
            # Lee el contenido del archivo
            contenido = archivo.read()
            
    return contenido
#-------------------------------------------------
```
